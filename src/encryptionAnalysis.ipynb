```python
import RSA
import time
import matplotlib.pyplot as plt
```

```python
plainText = RSA.processInput("Hello World!") # Known plaintext
n_bits = []
times = []
for numberOfBits in range(8, 1025, 2):
    publicKey, privateKey = RSA.generateKeys(numberOfBits)
    numberOfGroups = len(plainText) // RSA.GROUP_SIZE
    
    start = time.time()
    index = 0
    result = ""
    for i in range(numberOfGroups):
        C = RSA.encrypt(plainText[index: index+RSA.GROUP_SIZE], publicKey)
        index += RSA.GROUP_SIZE
        M = RSA.decrypt(int(C), privateKey)
        result += M
    end = time.time()
    
    times.append(end - start)
    n_bits.append(numberOfBits)
```

```python
plt.plot(n_bits, times)
plt.xlabel("Number of Bits")
plt.ylabel("Time (seconds)")
plt.title("Encryption/Decryption time vs Number of bits")
plt.show()
```
